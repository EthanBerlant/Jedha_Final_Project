```python
import pandas as pd
import numpy as np
```

```python
dat = pd.read_csv("stockData.csv")
```

```python
dat.describe()
```

```python
dat.columns = ['Date', 'SPX', 'TSLA', 'AAPL', 'NKE', 'AMZN', 'WMT', 'KO', 'Gold', 'Oil', 'UST10', 'UST2']
```

```python
dat.head()
```

```python
dat.dtypes
```

```python
dat['Date'] = pd.to_datetime(dat['Date'])
```

```python
#plot some charts
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
fig,ax = plt.subplots()
fig.set_size_inches (13,6)
plt.ylabel("Stock price")
ax = sns.lineplot (x="Date",y="SPX",data=dat,label="SPX")
ax1 = sns.lineplot (x="Date",y="TSLA",data=dat,label="TSLA")
ax2 = sns.lineplot (x="Date",y="AAPL",data=dat,label="AAPL")
ax3 = sns.lineplot (x="Date",y="NKE",data=dat,label="NKE")
ax4 = sns.lineplot (x="Date",y="AMZN",data=dat,label="AMZN")
ax5 = sns.lineplot (x="Date",y="WMT",data=dat,label="WMT")
ax6 = sns.lineplot (x="Date",y="KO",data=dat,label="CocaCola")
```

```python
fig,ax = plt.subplots()
fig.set_size_inches (13,6)
plt.ylabel("Other metrics")
ax = sns.lineplot (x="Date",y="SPX",data=dat,label="SPX")
ax1 = sns.lineplot (x="Date",y="Oil",data=dat,label="Oil")
ax2 = sns.lineplot (x="Date",y="UST10",data=dat,label="UST10")
ax3 = sns.lineplot (x="Date",y="UST2",data=dat,label="UST2")
ax3 = sns.lineplot (x="Date",y="Gold",data=dat,label="Gold")
```
